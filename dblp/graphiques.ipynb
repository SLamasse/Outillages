# <span style="color:#3db8ab">Aggrégation</span>

C'est semblable aux GRUP BY en SQL.
On peut utiliser deux fonctions :
* 'groupby'
* 'aggregate'

```python
#Calculer la population totale de chaque département

#Pour un département :
#data[data.dep==6].pop2012.sum()

#Avec un boucle for
for d in data.dep.unique():           #Renvoit les valeurs uniques dans la colonne, c'est comme un distinct en SQL.
    dep.courant = data[data.dep==d]   #Je sélectionne les villes du département d
    total = dep.courant.pop2012.sum() #Je calcule le total de sa population.
    print(d, total)
```

```python
#On peut faire la même chose, plus simplement, grâce au groupby :
deps = data.groupby("dep") #C'est comme si on faisait Select dep, ...from data group by dep
deps
```

**deps représente la liste des groupes des départements. Chaque groupe contient les villes ayant un même code de département.**

```python
#Pour visualiser le deps, on va l'afficher comme suit :
deps.groups
#C'est un dictionnaire dont les clefs sont les différentes valeurs de la colonne dep.
#Et la valeur associée à chaque code département contient les numéros des lignes de ce département.
#Donc le dép 1 est composé des lignes 0 à ...
```

**On peut appliquer des fonctions sur deps. Elles seront exécutées pour chacun des groupes dont il est formé.**

```python
deps['pop2012'].sum() #Pour chaque groupe de département, il fait la somme de pop2012.
#C'est l'équivalent de SELECT dep, sum(pop2012) FROM data GROUP BY dep
```

```python
#Si je cherche pour trois années, je peux passer une liste de colonnes :
deps[['pop2012','pop2010','pop1999']].sum()
```

**Si je veux des choses plus sophistiquées :**

```python
deps.aggregate({'pop1999':['min','max','mean','sum']
               ,'pop2010':['mean','sum']
               ,'pop2012':['min','max','sum']       
               })
```
